# Pandas and Plotly (a detour into Musk tweets)

### Import tweets

This code cell creates a list from some of Elon Musk's tweets. 

*You're not expected to completely understand this code yet. It's only here to give us interesting data*.

```python
import pandas

raw_tweets = pandas.read_csv('https://raw.githubusercontent.com/sim-mautner/cs1010-21t2/main/data/elonmusk.csv')
elon_tweets = list(raw_tweets.tweet)
```

### Questions

How many tweets are there in our dataset?

```python
len(elon_tweets)
```

In what tweets has Elon Musk said "simulation"?

```python
simulation_tweets = []
for tweet in elon_tweets:
  if "simulation" in tweet:
    simulation_tweets.append(tweet)

simulation_tweets
```

What emoji does Elon Musk use?

```python
!pip install emoji
import emoji

elon_emoji = []
for tweet in elon_tweets:
  for character in tweet:
    if character in emoji.UNICODE_EMOJI_ALIAS_ENGLISH:
      elon_emoji.append(character)

#elon_emoji
```

Can we build a dictionary counting how often Elon Musk uses each Emoji?

```python
emoji_count = {}

'''
smiley: 6
heart: 7
'''

for an_emoji in elon_emoji:
  # If it's in our dictionary, add one to the count
  # otherwise, add it to our dictionary with count = 1
  if an_emoji in emoji_count:
    emoji_count[an_emoji] = emoji_count[an_emoji] + 1
  else:
    emoji_count[an_emoji] = 1

emoji_count
```

Can we plot it?

```python
!pip install plotly
import plotly.express

plotly.express.bar(x=list(emoji_count.keys()), y=list(emoji_count.values()))
```

What about only the most frequently used ones?

```python
frequent_emojis = {}

for an_emoji in emoji_count:
  if emoji_count[an_emoji] > 10:
    frequent_emojis[an_emoji] = emoji_count[an_emoji]

plotly.express.bar(x=list(frequent_emojis.keys()), y=list(frequent_emojis.values()))
```

# Data Types: None

None is a special value in Python we can use to represent values that are unknown.

**Important Info**:

* The type of `None` is `NoneType`. It is the only value of that type.
* Python has special support for dealing with `None` values in `if` statements.

Write a function `maximum(numbers)` that finds the maximum number in a list.

```python
def maximum(numbers):
  max = None
  for num in numbers:
    # If we are at the first value, it's the max
    if max == None:
      max = num
    # If we are looking at a number bigger than max, it's the max
    elif num > max:
      max = num
  return max
```

```python
assert(maximum([1,2,3]) == 3)
assert(maximum([0,0,0]) == 0)
assert(maximum([]) == None)
assert(maximum([-1,-2,-3]) == -1)
```

# Exceptions

When things go wrong...

* An exception is an action that disrupts
the normal flow of a program.
* This action
is often representative of an error being
thrown.
* Exceptions are ways that we can
elegantly recover from errors.

Where have we seen exceptions before?

```python
print("hello" + 5)
print("that didn't work")
```

Can we recover from this more elegantly? (And even keep going with our program?)

```python
try:
  print("hello" + 5)
  print("still in the try - that didn't work")
except:
  print("in except because that didn't work")
```

Can we print out details of the exception?

```python
try:
  print("hello" + 5)
  print("still in the try - that didn't work")
except Exception as e:
  print("in except because that didn't work")
  print(f"Specifically, what went wrong is: {e}.")
```

What happens if we are deep inside functions?

```python
def level_one():
  level_two()

def level_two():
  level_three()

def level_three():
  print("hello" + 5)

level_one()
```

```python
def level_one():
  level_two()

def level_two():
  level_three()

def level_three():
  print("hello" + 5)

try:
  level_one()
except:
  print("Didn't work")
```

```python
def level_one():
  level_two()

def level_two():
  try:
    level_three()
  except Exception as e:
    print("That didn't work")
  print("But that's okay. Because we handled the exception.")

def level_three():
  print("hello" + 5)

level_one()
```

`else` and `finally`

* `else` runs if no exception occurred
* `finally` runs always (even if there's a `return` in the middle of the `except` block or `else` block)
* code after the `try-except` block runs as long as:
  * there are no un-handled exceptions
  * there are no previous return statements

```python
def level_one():
  level_two()

def level_two():
  try:
    level_three()
  except Exception as e:
    print("That didn't work")
    return
  else:
    print("No exception happened.")
  finally:
    print("But that's okay. This line will always happen.")
    
  print("Will we get here?")

def level_three():
  #print("hello" + 5)
  print("hello")

level_one()
```
